```python
houselist = []
for i in range(1,101):
    url = "https://bj.lianjia.com/zufang/pg" + str(i) + "/"
    raw_html = requests.get(url).content
    soup_doc = BeautifulSoup(raw_html, "html.parser") 
    results = soup_doc.find_all(class_='info-panel')    
    print("a new page")
    for result in results:
        housedic = { }
        housedic['name'] = result.find('h2').text
        housedic['url'] = result.find('h2').find('a')['href']
        housedic['building'] = result.find('div').find('span',  attrs={'class': 'region'}).string
        housedic['zoneinfo'] = result.find('div').find('span',  attrs={'class': 'zone'}).text
        housedic['area'] = result.find('div').find('span',  attrs={'class': 'meters'}).text
        housedic['direction'] = result.find('div').find('span',  attrs={'class': 'meters'}).next_sibling.string

        housedic['zone'] = result.find('div').find('div', attrs = {'class': 'con'}).text
        try:
            housedic['subway'] = result.find('div').find('div', attrs = {'class': 'chanquan'}).find('span', attrs={'class':'fang-subway-ex'}).string

            housedic['visiting'] = result.find('div').find('div', attrs = {'class': 'chanquan'}).find('span', attrs={'class':'haskey-ex'}).string
            housedic['decoration'] = result.find('div').find('div', attrs = {'class': 'chanquan'}).find('span', attrs={'class':'decoration-ex'}).string
            housedic['heat'] = result.find('div').find('div', attrs = {'class': 'chanquan'}).find('span', attrs={'class':'heating-ex'}).string
        except:
            housedic['subway'] = ''
            housedic['visiting'] = ''
            housedic['decoration'] =''
            housedic['heat'] =''

        housedic['price'] = result.find('div', attrs={'class':'col-3'}).find('div', attrs = {'class': 'price'}).find('span', attrs = {'class':'num'}).string    
        houselist.append(housedic)
    
houselist    
```
