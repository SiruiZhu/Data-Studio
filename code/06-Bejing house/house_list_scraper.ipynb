```python
def get_houseinfo(row):
    try:
        raw_html = requests.get(row['url']).content
        soup_doc = BeautifulSoup(raw_html, "html.parser") 
        type_layout = soup_doc.find(class_='zf-room').find('p').next_sibling.text
        neighborhood = soup_doc.find(class_='zf-room').find_all('a')[2].text
        location = soup_doc.find(class_='zf-room').find_all('a')[3].text
        layout= soup_doc.find('div', attrs={'class':'introContent'}).find_all('span', attrs={'class':'label'})[0].next_sibling
        heat_con = soup_doc.find('div', attrs={'class':'introContent'}).find_all('span', attrs={'class':'label'})[3].next_sibling
        print("heat_condition is", heat_con)

        return pd.Series({
            'type_layout':type_layout,
            'neighborhood': neighborhood,
            'location':location,        
            'layout':layout,
            'heat_con':heat_con       

            })
    except:
        return pd.Series({}) 
```

```python
df_house_total = df_house.apply(get_houseinfo, axis=1).join(df_house)
df_house_total
```
