```python
df_chinavisa = df_china[df_china['country'].isin(["China - mainland", "China - Taiwan", "Hong Kong S.A.R."])]
df_chinavisa
```

```python
df_chinaf1 = df_chinavisa.groupby('year').F1.sum().reset_index(name="F1 Visa issued to Chinese")
df_chinaf1
```

```python
##final dataframe
df_info = pd.merge(df_chinaf1, df_total, on=['year'])
df_info = df_info.drop(columns=['country'])
df_info['year'] = df_info.year.astype(int)
df_info
```

```python
df_info['pct_f1'] =  df_info['F1 Visa issued to Chinese'] / df_info['F1'] * 100
df_info
```
