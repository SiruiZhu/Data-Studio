```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import glob
```

```python
country2013 = pd.read_csv("country2013.csv",  encoding = "ISO-8859-1")
country2013.head()
```

```python
country2013 = country2013[['Unnamed: 0', 'USA', 'Brazil', 'Mexico', 'Germany', 'Colombia']]
country2013['year'] = 2013
country2013.head()
```

```python
country2014 = pd.read_csv("country2014.csv",  encoding = "ISO-8859-1")
country2014.head()
```

```python
country2014 = country2014[['Unnamed: 0', 'USA', 'Brazil', 'Mexico', 'Germany', 'Colombia']]
country2014['year'] = 2014
country2014
```

```python
country2015 = pd.read_csv("country2015.csv",  encoding = "ISO-8859-1")
country2015.head()
```

```python
country2015 = country2015[['Unnamed: 0', 'USA', 'Brazil', 'Mexico', 'Germany', 'Colombia']]
country2015['year'] = 2015
country2015
```

```python
frames = [country2013,country2014,country2015]
```

```python
countrytotal = pd.concat(frames)
countrytotal.head(40)
```

```python
country_total = countrytotal[countrytotal['Unnamed: 0'].isin(['Total Face & Head Procedures', 'Total Breast Procedures', 'Total Body & Extremities Procedures', 'Total Non-Surgical Procedures'])]
```

```python
country_total['USA'] = country_total['USA'].str.replace(',','')
```

```python
country_total['Brazil'] = country_total['Brazil'].str.replace(',','')
```

```python
country_total['Mexico'] = country_total['Mexico'].str.replace(',','')
```

```python
country_total['Germany'] = country_total['Germany'].str.replace(',','')
country_total['Colombia'] = country_total['Colombia'].str.replace(',','')
```

```python
country_total
```

```python
country_total.face = country_total[country_total['Unnamed: 0'] == 'Total Face & Head Procedures']
country_total.face.sort_values(by='year', ascending=True)
```

```python
country_total.face['USA'] = country_total.face['USA'].astype(float)
country_total.face['Brazil'] = country_total.face['Brazil'].astype(float)
country_total.face['Mexico'] = country_total.face['Mexico'].astype(float)
country_total.face['Germany'] = country_total.face['Germany'].astype(float)
country_total.face['Colombia'] = country_total.face['Colombia'].astype(float)
```

```python
ax5 = country_total.face.plot(x='year', kind='bar')
ax5.set_title("Who Cares Their Face and Head Most?", fontsize=20, pad=50, loc ='center', weight='bold')
ax5.grid(False, axis='x')
ax5.xaxis.set_ticks_position('bottom')
ax5.tick_params(which='major', left=False, bottom=False, top=False, grid_alpha=0.3)
ax5.yaxis.label.set_visible(False)
ax5.xaxis.label.set_visible(False)

plt.xticks(rotation=0)
ax5.legend(
           ncol=5, 
           loc=9, #upper center
           bbox_to_anchor=(0.5, 0.9),
           frameon=False,
           columnspacing=3,
           labelspacing=1,
           bbox_transform=plt.gcf().transFigure)

from matplotlib.ticker import FuncFormatter
def with_commas(value, pos):
    return "{:,}".format(int(value))
formatter = FuncFormatter(with_commas)

def with_words(value, pos):
    millions = value / 1000000
    if millions == 0.5:
        return "{} Millions". format(millions)
    elif millions == 0.0:
        return 0
    else:
        return millions

formatter = FuncFormatter(with_words)
ax5.yaxis.set_major_formatter(formatter)
plt.tight_layout()
plt.savefig('output5.pdf')
plt.savefig('output5.svg')
```
