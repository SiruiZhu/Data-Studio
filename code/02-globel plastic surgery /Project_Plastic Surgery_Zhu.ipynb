```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import glob
```

```python
filenames = glob.glob("total-*.csv")
filenames
```

```python
dataframe = []
for filename in filenames:
    df = pd.read_csv(filename)
    df['filename'] = filename   
    df['type'] = df['filename'].str.extract(r"-(.*).csv")
    
    dataframe.append(df)

global2016 = pd.concat(dataframe, ignore_index=True, sort=True)

global2016['RANK'] = global2016['RANK'].fillna(global2016['rank'])
global2016.drop(columns=['rank', 'filename'], inplace=True)
global2016.head(100)
```

```python
global2016['RANK'] = global2016['RANK'].astype(object)
global2016.head(100)
```

```python
global2016['% OF WORLD-WIDE TOTAL'] = global2016['% OF WORLD-WIDE TOTAL'].str.extract(r"(\d+.*)%")
global2016.head(100)
```

```python
global2016.dtypes
```

```python
global2016['PROCEDURE'] = global2016['PROCEDURE'].str.replace(',','')
global2016['PROCEDURE'].astype(int)
```

```python
global2016 = global2016[['RANK', 'COUNTRY', 'PROCEDURE', 'type', '% OF WORLD-WIDE TOTAL']]
global2016.head()
```

```python
brazil = global2016[global2016['COUNTRY'] == 'Brazil']
brazil
```

```python
global2016_new = global2016.sort_values(by="COUNTRY")
global2016_new.head(40)
```

```python
global2016_wide = global2016_new.pivot('COUNTRY', 'type', 'PROCEDURE')
global2016_wide
```
