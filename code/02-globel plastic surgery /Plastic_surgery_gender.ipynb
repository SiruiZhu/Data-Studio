```python
import pandas as pd
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
import glob
```

```python
filenames = glob.glob("gender-*.csv")
filenames
```

```python
dataframe = []
for filename in filenames:
    df = pd.read_csv(filename, encoding = "ISO-8859-1")
    df['filename'] = filename   
    df['year'] = df['filename'].str.extract(r"-(\d\d\d\d).csv")
    
    dataframe.append(df)

genders = pd.concat(dataframe, ignore_index=True, sort=True)

#global2016['RANK'] = global2016['RANK'].fillna(global2016['rank'])
#global2016.drop(columns=['RANK', 'filename'], inplace=True)
genders.head(100)
```

```python
genders = genders.sort_values(by='year')
genders.head()
```

```python
genders.replace("dna", "NaN", inplace=True)
genders.head(30)
```

```python
genders['FEMALE Number of Procedures'] = genders['FEMALE Number of Procedures'].str.replace(',','')
genders['MALE Number of Procedures'] = genders['MALE Number of Procedures'].str.replace(",", "")
```
